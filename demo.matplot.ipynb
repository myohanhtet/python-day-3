```python
x = ['python', 'java', 'c++', 'javascript', 'ruby', 'php']
y = [10, 20, 10, 40, 30, 20]
c = ['red', 'blue', 'green', 'orange', 'purple', 'brown']
explode = (0.1, 0, 0, 0, 0, 0)  # only "explode" the 1st slice (i.e. 'python')
plt.pie(y, labels=x, colors=c, autopct='%1.1f%%',explode=explode, shadow=True, startangle=140)
plt.title('Programming Languages Popularity')
# plt.xlabel('Languages')
# plt.ylabel('Popularity')
plt.legend(x, loc='upper right')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
